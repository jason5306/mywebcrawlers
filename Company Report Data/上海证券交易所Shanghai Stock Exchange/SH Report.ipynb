```python
def SH_Dowload2TXT(code):
    import datetime
    import wget
    import os


    report_type = '年报'
    begin = '2018-01-01'
    end = '2021-09-20'

    begin = datetime.datetime.strptime(begin,'%Y-%m-%d').date()
    end = datetime.datetime.strptime(end,'%Y-%m-%d').date()

    from shreport import SH

    cookies = {"Cookie": 'VISITED_MENU=%5B%228312%22%5D; sseMenuSpecial=8311; yfx_f_l_v_t_10000042=f_t_1628514766478__r_t_1628514766478__v_t_1628514875695__r_c_0; yfx_c_g_u_id_10000042=_ck21080921124616006732345236521'}
    sh = SH(cookies)


    #获取浦发银行披露信息
    df = sh.disclosure(code)

    #存储数据
    #df.to_excel('600000.xlsx')

    #前5条信息
    df.head()

    # 选择一段时期和报告类型下载
    selected_df = df[(df['type'] == report_type)]

    os.makedirs("/Users/jiacheng/Documents/MacBookPro/Academic/iMep/KeyWord/SH_report/" + code)

    for index, row in selected_df.iterrows():
        current_row_date = row['date']
        current_row_date = datetime.datetime.strptime(current_row_date,'%Y-%m-%d').date()
        if current_row_date > begin and current_row_date < end:
            file_url = row['pdf']
            wget.download(file_url,"/Users/jiacheng/Documents/MacBookPro/Academic/iMep/KeyWord/SH_report/" + code)
            print("Downloading...",row['company'], row['type'] ,current_row_date)
    print("Done")

    filePath = "/Users/jiacheng/Documents/MacBookPro/Academic/iMep/KeyWord/SH_report/" + code
    # Get remaining file names
    fileNames = os.listdir(filePath)

    # Change extension to .pdf and exclude non_pdf files
    real_pdf_file_names = []
    for fileName in fileNames:
        extension = fileName.split(".")[1] 
        if extension == 'pdf':
            real_pdf_file_names.append(fileName)
        if extension == 'PDF':
            new_fileName = fileName.split(".")[0] + ".pdf"
            os.rename(filePath + '/' + fileName, filePath + '/' + new_fileName)
            real_pdf_file_names.append(new_fileName)
    print(real_pdf_file_names)

    report_name = real_pdf_file_names[0].split('_')[0]
    result_file = '/Users/jiacheng/Documents/MacBookPro/Academic/iMep/KeyWord/TXTs/SH_TXT/' + report_name +'.txt'
    file = open(result_file,"a")

    for name in real_pdf_file_names:
        convert_file = filePath + '/' + name.split('.')[0] + '.txt'
        pdf_file =filePath + '/' + name
        cmd = 'pdf2txt.py ' + pdf_file + ' -o ' + convert_file
        os.system(cmd)
        print(convert_file + '...Converted')
        merge_txt(result_file, convert_file)
        print(convert_file + '...Merged')
        os.remove(convert_file)
        os.remove(pdf_file)
```

```python
def merge_txt(file1, file2):
    f1 = open(file1, "a+", encoding='utf-8')
    with open(file2, 'r', encoding = 'utf-8') as f2:
        f1.write('\n')
        for i in f2:
            f1.write(i)
```
