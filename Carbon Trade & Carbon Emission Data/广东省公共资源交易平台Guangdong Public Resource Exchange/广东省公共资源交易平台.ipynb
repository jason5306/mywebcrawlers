# 排污权交易目录

```python
driver = webdriver.Chrome('../chromedriver')
driver.get('https://ygp.gdzwfw.gov.cn/ggzy-portal/#/44/jygg')
time.sleep(3)   
```

```python
driver.find_element_by_class_name('toggle-content').find_element_by_xpath("//li[text()='排污权']").click()
header = ['标题','发布时间']
result = []
total_page = driver.find_element_by_class_name('gd-pagination-to').find_element_by_tag_name('span').text.split('/')[1].strip()
total_page

for i in range(int(total_page)):
    print("page:" + str(i+1))
    data_list = driver.find_element_by_class_name('jygg-list').find_elements_by_tag_name('li')
    for item in data_list:
        title = item.find_element_by_tag_name('h2').text
        date = item.find_element_by_class_name('br').text.split('：')[1]
        result.append([title, date])    
#     next page
    try:
        driver.find_element_by_css_selector(".gd-icon.icon-right.only-icon").find_element_by_xpath('..').click()
    except Exception:
        print('---End---')
        break
    time.sleep(3) 

file_name = "广东省公共资源交易平台-排污权交易目录-" + time.strftime("%Y-%m-%d", time.localtime()) + ".csv"
pd.DataFrame(result,columns=header).to_csv(file_name)    
```

# 结果公告

```python
driver = webdriver.Chrome('../chromedriver')
driver.get('https://ygp.gdzwfw.gov.cn/ggzy-portal/#/44/jygg')
time.sleep(3)  
# go to correct tab
driver.find_element_by_class_name('toggle-content').find_element_by_xpath("//li[text()='排污权']").click()
time.sleep(1)
driver.find_element_by_css_selector('.label-secondType.pr-5.toggle').find_element_by_xpath("//li[text()='结果公告']").click()
```

```python
df_result = pd.DataFrame()
total_page = driver.find_element_by_class_name('gd-pagination-to').find_element_by_tag_name('span').text.split('/')[1].strip()
print(total_page)

for i in tqdm(range(int(total_page))):
#     print("page:" + str(i+1))
    while True:
        try:
            ele_list = driver.find_element_by_class_name("jygg-list").find_elements_by_tag_name('li')
        except Exception:
            time.sleep(3)
        else:
            if ele_list != "":
                break
            else:
                print('Page reload ... ')
                time.sleep(3)                

    for ele in ele_list:
#         print(ele.find_element_by_tag_name('h2').text)
        ele.click()
        time.sleep(3)  

        # switch driver to the new tab
        handles = driver.window_handles
        driver.switch_to.window(handles[1])

        result = {}
        while True:
            try:
                table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
            except Exception:
                print('Page reload ... ')
                driver.refresh()
                time.sleep(3)
            else:
                if table_row != "":
                    break
                else:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)

        table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
        for tr in table_row:
            td = tr.text.split('\n')
            if len(td) == 2:
                result[td[0]] = [td[1]]
            if len(td) == 4:
                result[td[0]] = [td[1]]
                result[td[2]] = [td[3]]

        # switch driver to iframe        
        iframe_id = driver.find_element_by_tag_name('iframe').get_attribute('id')
        driver.switch_to.frame(iframe_id)

        while True:
            try:
                iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
            except Exception:
                print('Page reload ... ')
                driver.refresh()
                time.sleep(3)
            else:
                if iframe_table_row != "":
                    break
                else:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)        
        
        iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
        for tr in iframe_table_row:
            td = tr.find_elements_by_tag_name('td')
            if len(td) == 2:
                result[td[0].text.strip('\n')] = [td[1].text.strip('\n')]
            if len(td) == 3:
                result[td[1].text.strip('\n')] = [td[2].text.strip('\n')]
            if len(td) == 4:
                result[td[0].text.strip('\n')] = [td[1].text.strip('\n')]
                result[td[2].text.strip('\n')] = [td[3].text.strip('\n')]

        df_temp = pd.DataFrame(result)
        df_result = pd.concat([df_result, df_temp], axis=0, ignore_index=True) 


        driver.close()
        # switch back to main tab
        handles = driver.window_handles
        driver.switch_to.window(handles[0]) 
        
    # next page
    try:
        driver.find_element_by_css_selector(".gd-icon.icon-right.only-icon").find_element_by_xpath('..').click()
    except Exception:
        print('---End---')
        break
    time.sleep(3)        
```

```python
file_name = "广东省公共资源交易平台-排污权-结果公告" + time.strftime("%Y-%m-%d", time.localtime()) + ".csv"
df_result.to_csv(file_name)
```

# 交易公告

```python
driver = webdriver.Chrome('../chromedriver')
driver.get('https://ygp.gdzwfw.gov.cn/ggzy-portal/#/44/jygg')
time.sleep(3)  
# go to correct tab
driver.find_element_by_class_name('toggle-content').find_element_by_xpath("//li[text()='排污权']").click()
time.sleep(1)
driver.find_element_by_css_selector('.label-secondType.pr-5.toggle').find_element_by_xpath("//li[text()='交易公告']").click()
```

```python
df_result = pd.DataFrame()
total_page = driver.find_element_by_class_name('gd-pagination-to').find_element_by_tag_name('span').text.split('/')[1].strip()
print(total_page)

for i in tqdm(range(int(total_page))):
#     print("page:" + str(i+1))
    while True:
        try:
            ele_list = driver.find_element_by_class_name("jygg-list").find_elements_by_tag_name('li')
        except Exception:
            time.sleep(3)
        else:
            if ele_list != "":
                break
            else:
                print('Page reload ... ')
                time.sleep(3)                

    for ele in ele_list:
        if '交易公告' in ele.find_element_by_tag_name('h2').text:
            ele.click()
            time.sleep(3)  

            # switch driver to the new tab
            handles = driver.window_handles
            driver.switch_to.window(handles[1])  
            result = {}
            while True:
                try:
                    table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
                except Exception:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)
                else:
                    if table_row != "":
                        break
                    else:
                        print('Page reload ... ')
                        driver.refresh()
                        time.sleep(3)

            table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
            for tr in table_row:
                td = tr.text.split('\n')
                if len(td) == 2:
                    result[td[0]] = [td[1]]
                if len(td) == 4:
                    result[td[0]] = [td[1]]
                    result[td[2]] = [td[3]]   

            # switch driver to iframe        
            iframe_id = driver.find_element_by_tag_name('iframe').get_attribute('id')
            driver.switch_to.frame(iframe_id)

            while True:
                try:
                    iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
                except Exception:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)
                else:
                    if iframe_table_row != "":
                        break
                    else:
                        print('Page reload ... ')
                        driver.refresh()
                        time.sleep(3)        

            iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
            trade_info = []
            for tr in iframe_table_row:
                td = tr.find_elements_by_tag_name('td')
                td_length = 0
                for item in td:
                    if item.text:
                        td_length += 1
                if td_length == 2:
                    result[td[0].text.strip()] = [td[1].text.strip()]
                if td_length > 2:
                    trade_info.append(tr)

            trade_info_header = []
            for item in trade_info[0].find_elements_by_tag_name('td'):
                if item.text:
                    trade_info_header.append(item.text)

            for j in range(1,len(trade_info)):    
                td = trade_info[j].find_elements_by_tag_name('td')
                for i in range(0,len(td)):
                    if td[i].text:
                        result[trade_info_header[i]+'_'+str(j)] = [td[i].text]    


        elif '出让公告' in ele.find_element_by_tag_name('h2').text:
            ele.click()
            time.sleep(3)  

            # switch driver to the new tab
            handles = driver.window_handles
            driver.switch_to.window(handles[1])  
            result = {}
            while True:
                try:
                    table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
                except Exception:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)
                else:
                    if table_row != "":
                        break
                    else:
                        print('Page reload ... ')
                        driver.refresh()
                        time.sleep(3)

            table_row = driver.find_element_by_class_name("mailTable").find_elements_by_tag_name('tr')
            for tr in table_row:
                td = tr.text.split('\n')
                if len(td) == 2:
                    result[td[0]] = [td[1]]
                if len(td) == 4:
                    result[td[0]] = [td[1]]
                    result[td[2]] = [td[3]]      

            # switch driver to iframe        
            iframe_id = driver.find_element_by_tag_name('iframe').get_attribute('id')
            driver.switch_to.frame(iframe_id)

            while True:
                try:
                    iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
                except Exception:
                    print('Page reload ... ')
                    driver.refresh()
                    time.sleep(3)
                else:
                    if iframe_table_row != "":
                        break
                    else:
                        print('Page reload ... ')
                        driver.refresh()
                        time.sleep(3)        

            iframe_table_row = driver.find_element_by_tag_name('table').find_elements_by_tag_name('tr')
            for tr in iframe_table_row:
                td = tr.find_elements_by_tag_name('td')
                if len(td) == 2:
                    result[td[0].text.strip('\n')] = [td[1].text.strip('\n')]                                            

        df_temp = pd.DataFrame(result)
        df_result = pd.concat([df_result, df_temp], axis=0, ignore_index=True) 


        driver.close()
        # switch back to main tab
        handles = driver.window_handles
        driver.switch_to.window(handles[0])

    # next page
    try:
        driver.find_element_by_css_selector(".gd-icon.icon-right.only-icon").find_element_by_xpath('..').click()
    except Exception:
        print('---End---')
        break
    time.sleep(3) 



```

```python
file_name = "广东省公共资源交易平台-排污权-交易公告" + time.strftime("%Y-%m-%d", time.localtime()) + ".csv"
df_result.to_csv(file_name)
```
