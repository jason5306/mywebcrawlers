```python
import requests as rq
from lxml import etree
import re
from pandas import read_csv, DataFrame, set_option

def get_html(url):
    HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}

    # test_url = "https://www.cneeex.com/c/2021-07-16/491500.shtml"
    test_url = url
    test_res = rq.get(test_url, headers = HEADERS)        #Get方式获取网页数据
    test_res.encoding = 'utf-8' # 指定编码
    test_html = test_res.text
    rq.adapters.DEFAULT_RETRIES = 5
    #print(test_html)
    return test_html

```

```python
li1 = []
li2 = []
li3 = []
li4 = []
li5 = []

pages = 2000

for i in range(1,pages):

    print('page:' + str(i))
    url = "http://www.hbets.cn/list/11.html?page=" + str(i)
    html = get_html(url)

    start = '<ul class="cont clearfix">'
    end = '</li>'
    index_start = html.find(start)

    while(True):

        index_li1 = html.find('li1', index_start)
        index_end = html.find(end, index_li1)
        text1 = html[index_li1+5:index_end]
        print(text1)
        li1.append(text1)

        index_li2 = html.find('li2', index_end)
        index_end = html.find(end, index_li2)
        text2 = html[index_li2+5:index_end]
        print(text2)
        li2.append(text2)

        index_li3 = html.find('li3', index_end)
        index_end = html.find(end, index_li3)
        text3 = html[index_li3+23:index_end]
        print(text3)
        li3.append(text3)

        index_li4 = html.find('li4', index_end)
        index_end = html.find(end, index_li4)
        text4 = html[index_li4+23:index_end]
        print(text4)
        li4.append(text4)

        index_li5 = html.find('li5', index_end)
        index_end = html.find(end, index_li5)
        text5 = html[index_li5+5:index_end]
        print(text5)
        li5.append(text5)

        index_start = html.find(start, index_end)
        if index_start == -1:
            break
```

```python
df = DataFrame((zip(li1,li2,li3,li4,li5)), columns = ['试点地区','成交价','当日成交量（吨）','当日成交额（元）','日期'])
df
```
